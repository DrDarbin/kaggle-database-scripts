```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

Users = pd.read_csv('../input/Users.csv').rename(columns = {'Id': 'UsersId'})
Kernels = pd.read_csv('../input/Kernels.csv').rename(columns = {'Id': 'KernelsId'})
KernelVersions = pd.read_csv('../input/KernelVersions.csv').rename(columns = {'Id': 'KernelVersionsId'})
KernelVotes = pd.read_csv('../input/KernelVotes.csv').rename(columns = {'Id': 'KernelVotesId'})
```

```python
# show all columns ad only 200 rows in tables
pd.set_option('display.max_columns', None)
pd.options.display.max_rows = 600
```

```python
# kernels upvoted by a user
user_name = 'dandrocec'
user_id = float(Users[Users.UserName == user_name]['UsersId'])
KernelVotes[KernelVotes.UserId == user_id].sort_values(by = ['VoteDate'], ascending = True)
```

```python
# get a link to a chosen kernel
kernel_version_id = 393887
kernel_id = float(str(KernelVersions[KernelVersions.index == kernel_version_id]['KernelId']).split("    ")[1].split('\nName:')[0])
kernel_url_slug = Kernels[Kernels.index == kernel_id]['CurrentUrlSlug']
kernel_author_id = Kernels[Kernels.index == kernel_id]['AuthorUserId']
kernel_author_name = Users[Users.UsersId.isin(kernel_author_id)]['UserName']
author = str(kernel_author_name).split("    ")[1].split('\nName:')[0]
slug = str(kernel_url_slug).split("    ")[1].split('\nName:')[0]
url = "https://www.kaggle.com" + "/" + author + "/" + slug
url
```
